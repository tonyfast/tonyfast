# discovering assistive technology throught the thingiverse api

```python
%%
there are two ways to search

1. search by `term`

        __import__("dotenv").load_dotenv()
        params = dict(access_token=os.environ["THINGIVERSE_ACCESS_TOKEN"])
        import requests    
        term = requests.get(
            "https://api.thingiverse.com/search/assistive+technology",
            params | dict(per_page=30, page=1, type="thing")
        )

1. search by `tag`

        tag = requests.get("https://api.thingiverse.com/search/assistive_technology/tag", params)
```

```python
%%
## the thingiverse data

    terms, tags = DataFrame(term.json()), DataFrame(tag.json())    
    print(F"""{terms.loc[0, "total"]} terms and {tags.loc[0, "total"]} tags""")
    
    data = tags["hits"].series().set_index("id")
    data.tags = data.tags.explode().series().tag.groupby(pandas.Grouper(level=0)).agg(list)
    

{{data.T._repr_html_()}}

### the counts of all the tags collection for 30 samples

{{data.tags.explode().value_counts().to_frame("count").T.style._repr_html_()}}
```
