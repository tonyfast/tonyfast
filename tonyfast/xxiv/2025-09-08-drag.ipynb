````python
%%
<details><summary>boilerplate</summary>

    import pypugjs
    pug_process = compose_left(lambda x: pypugjs.process(x, **globals()), HTML)
    shell.tangle.parser.fence_methods["pug"] = "__main__:pug_process"

</details>


    html = do(display)\
```pug
section#x
    p.
        words you can drag #[br]
        & zoom & #[br]#[kbd Double Click]
form(name="transform")
    fieldset 
        legend controls
        p it is important to have separate controls for keyboard users
        label x
            input(type="number", name="cx", aria-controls="x", data-unit="px", value=0)
        label y 
            input(type="number", name="cy", aria-controls="x", data-unit="px", value=0)
        label scale
            input(type="number", name="scale", aria-controls="x", data-unit="", value=1)
style.
    :root {
        --w: 600px;
        --h: 300px;
    }
    section#x {
        display: block;
        position: relative;
        width: var(--w); 
        height: var(--h);
        p {
            position: absolute;
            top: 0;
            left: 0;
            transform-origin: var(--cx, calc(var(--w) /2)) var(--cy, calc(var(--h) /2));
            transform: 
                scale(var(--scale, 1))
                translate(var(--cx, calc(var(--w) /2)), var(--cy, calc(var(--h) /2)))
                translate(-50%, -50%);
        }
    }
script(type="module").
    import * as d3 from "https://cdn.jsdelivr.net/npm/d3@7/+esm"
    d3.select("#x").call(
        d3.drag().on("drag", (event)=>{
            console.log(event)
            d3.select("#x")
                .style("--cx", `${event.x}px`)
                .style("--cy", `${event.y}px`)
            document.forms.transform.cx.value = event.x
            document.forms.transform.cy.value = event.y
        })
    ).call(
        d3.zoom().on("zoom", (event)=>{
            d3.select("#x")
                .style("--scale", event.transform.k)
            document.forms.transform.scale.value = event.transform.k
        })
    )

script.
    Array.from(document.forms.transform.elements).forEach(
        (element) => {
            console.log(element)
            element.oninput = (event) => {
                console.log(event, document.getElementById(event.target.getAttribute("aria-controls")))
                document.getElementById(event.target.getAttribute("aria-controls"))
                    .style.setProperty("--"+event.target.name, 
                        event.target.value + event.target.dataset.unit)
        }
    })
    
```

<details><summary>raw html</summary>

```html
{{html}}
```

</details>
````
