````python
%%

```javascript
var devices = [];
function getDevices(deviceInfos){
    for (let i = 0; i !== deviceInfos.length; ++i) {
        const deviceInfo = deviceInfos[i];
        if (deviceInfo.kind === 'videoinput') {
          devices.push(deviceInfo);
        } else {
          //console.log('Some other kind of source/device: ', deviceInfo);
        }
    }
    return devices;
  }
var template = document.getElementById(`video`);
var target = document.getElementById(`videos`);

function changeSource(event) {
    console.log(event.target.value, event.target.value == "");
    var video = event.target.closest(`section`).querySelector(`video`);
    var videos = document.querySelectorAll(`video.webcam`);
    if (event.target.value == "") {
        video.srcObject = null;
        videos.forEach((video) => {video.srcObject = null;})
        return
    }
    navigator.mediaDevices.getUserMedia({ 
        video: {
            facingMode: `forward`,
            deviceId: {
                exact: event.target.value,
            }
        }, audio: false 
    }).then(stream => {videos.forEach((video) => {video.srcObject = stream;})}).catch(console.log);
}

function getContainers(devices) {
    devices.slice(0,1).map((device) => {
        if (device.deviceId == "") {
            return
        }
        var exists = document.getElementById(device.deviceId);
        if (exists !== null) {
            return
        }
        var container = template.content.cloneNode(true);
        container.querySelector(`section`).setAttribute(`id`, device.deviceId);
        var select = container.querySelector(`select`);
        select.onchange = changeSource;
        devices.map((d, i) => {
            var option = document.createElement(`option`);
            option.text = d.label || `video ${i}`;
            option.setAttribute(`value`, d.deviceId);
            select.appendChild(option);
        });
    
        target.appendChild(container);
    })
}

navigator.mediaDevices.getUserMedia({
    video: {deviceId: undefined}
}).then(
    (stream) => {
        navigator.mediaDevices.enumerateDevices().then(getDevices).then(getContainers)
    }
);
```
````
