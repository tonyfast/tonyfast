```python
%%
    I = 24
<ol class=circles style="--I: {{I}};">{% for i in range(I) %}<li style="--i: {{i}};">&nbsp;</li>{% endfor %}</ol>
```

```python
%%
<form name=circles>
<label>
          radius
          <input type=number name=radius value=2 >
</label>
                                            <br>
<label>
          amplitude
          <input type=number name=amplitude value=1 step=.5>
</label>
</form>
<script>
document.forms.circles.radius.onchange = (e) => {
    document.querySelectorAll(`.circles`).forEach((target)=>{target.style.setProperty(`--r`, `${e.target.value}rem`)})
}
document.forms.circles.amplitude.onchange = (e) => {
    document.querySelectorAll(`.circles`).forEach((target)=>{target.style.setProperty(`--amplitude`, `${e.target.value}rem`)})
}
</script>
```
