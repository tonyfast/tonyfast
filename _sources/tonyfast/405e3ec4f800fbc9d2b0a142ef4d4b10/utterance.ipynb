# A quick utteranc.es for IPython and Pyolite

```python
try:
    from IPython.display import *
    __main_name__ = "__main__"
except:
    from pyolite.display import *
    __main_name__ = "__console__"
```

```python
def utter(
    repo="utterance/utterances", 
    issue="1", 
    theme="preferred-color-scheme",
    style="social"
):
    gh_html = f"https://github.com/{repo}/issues/{issue}"
    gha = f"""<a target="about:blank" href="{gh_html}" """
    ghi = f"""<i class="fa fa-github"></i>"""
    shield = lambda field, label: f"""
        <img src="https://img.shields.io/github/issues/detail/{field}/{repo}/{issue}?style={style}{'&label=' + label if label else ''}" />
    """
    return HTML(f"""
    <div class="uttered">
        <h2>{gha}>@{repo}#{issue} {ghi}</a></h2>
        <blockquote>
            {gha}>
            {shield("title", "")}
            {shield("state", "is")}
            {shield("comments", "💬")}
            {shield("comments", "📆")}
            <code>{repo} {ghi}</code>
            </a>
        </blockquote>
        <script async src="https://utteranc.es/client.js" crossorigin="anonymous" 
            repo="{repo}" issue-number="{issue}" theme="{theme}">
        </script>
    </div>
    <style>
        .uttered {{ 
            max-height: 90vh; 
            overflow-y: auto; 
            flex: 1; 
            height: 100%; 
        }}
        .uttered h2 {{ padding: 2em 1em 0 1em; }}
        .uttered blockquote {{ 
            padding-left: 0;
            padding-top: 0.5em;
            margin-left: 0;
            position: absolute; 
            top: -0.25em; 
            right: 0.5em;
            left: 0;
            z-index: 1; 
            background-color: var(--jp-layout-color1); 
            padding: 0.25em; 
            border-radius: 0.25em;
            opacity: 0.8;
            margin: 0;
            transition: opacity 0.5s;
        }}
        .uttered blockquote:hover {{ 
            opacity: 1;
            transition: all 0.5s;
        }}
        .uttered blockquote i {{
            font-size: 1em;
        }}
        .uttered blockquote a {{ 
            padding: 0.25em;
            display: flex;
            flex-direction: row;
            align-items: center;
            justify-content: space-between;
        }}
        .uttered blockquote code {{
            border-radius: 0.5em;
        }}
    </style>
    """)
```

```python
if __name__ == __main_name__:
    display(utter())
```
